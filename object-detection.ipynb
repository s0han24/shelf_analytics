```python
dataset_dir = "/kaggle/input/sku110k-annotations/data_kaggle.yaml" # enter the dataset directory, the dataset should be in YOLO format
```

```python
# simply train YOLOv8 on the SKU110k dataset
from ultralytics import YOLO

model = YOLO("yolov8s.yaml").load("yolov8n.pt")

# Train the model
results = model.train(data=dataset_dir, epochs=50, imgsz=640, fraction=0.25)

val_results = model.val()
```
